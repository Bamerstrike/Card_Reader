```python
# This file is used to convert files of pictures or videos and rename them into the files YOLO needs.

import os

```

```python
directory = "playing-cards/img"
move_to_directory = "playing-card-detection/data/card_pictures"

# iterate through the directory playing-cards for card images
for filename in os.listdir(directory):
#     print (filename)
    # remove the beginning format of each file
    card = filename.replace("cards-[","")
    card = card[0:2]
#     print(card)
    card_suite = card[0:1].lower()
#     print(card_suite)
    card_value = card[1:2]
#     print(card_value)

    # turns the 0 cards into 10
    try:
        if int(card_value) == 0:
            card_value = str(10)
    except (ValueError):
        pass
    
    card = card_value+card_suite
#     print(card)

    picture_name = directory+"/"+filename
#     move_to_name = move_to_directory+"/"+card+"/"+card+".jpg"
    move_to_name = move_to_directory+"/"+card+".jpg"
    print(move_to_name)
    
    try:
        os.rename(picture_name, move_to_name)
    except(FileExistsError,FileNotFoundError):
        pass

    
    
    
    
```
